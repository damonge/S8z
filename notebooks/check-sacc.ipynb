```python
import numpy as np
import sacc

root = '/mnt/extraspace/gravityls_3/S8z/Cls/all_together_metacal_new/'
```

# $C_\ell$ (Check passed)

```python
s = sacc.Sacc.load_fits(root + 'DESwl.fits')
dtypes = s.get_data_types()
trcomb = s.get_tracer_combinations()
```

```python
cl = np.load(root + 'cl_all_no_noise.npz')
lbpw, cls =  cl['l'], cl['cls']
```

```python
for tr1, tr2 in trcomb:
    b1 = int(tr1[-1]) + 5
    b2 = int(tr2[-1]) + 5
    ibin1 = 2*int(tr1[-1]) + 5
    ibin2 = 2*int(tr2[-1]) + 5
    cov = np.load(root+'cov_new_fiducial/cov_s2222_b{0}{1}{0}{1}.npz'.format(b1, b2))['arr_0']
    cov = cov.reshape(lbpw.size, 4, lbpw.size, 4)
    for j, dt in enumerate(dtypes):
        if dt == 'cl_be':
            continue
        elif dt == 'cl_bb':
            ibin1 += 1  # As dtype in order in the list, cl_eb has summed 1 to ibin2 already
            ic1 = 3
        elif dt == 'cl_eb':
            ibin2 +=1
            ic1 = 1
        else:
            ic1 = 0
            
        slbpw, scls, scov, sind = s.get_ell_cl(dt, tr1, tr2, return_cov=True, return_ind=True)
        print('Cl', np.all(cls[ibin1, ibin2] == scls))
        
        print('Cov', np.max(np.abs(scov - cov[:, ic1, :, ic1])))
```

```python
scov = s.covariance.covmat
ic = {'cl_ee': 0,
      'cl_eb': 1,
      'cl_be': 2,
      'cl_bb': 3}
for i, (tra1, tra2) in enumerate(trcomb):
    b1 = int(tra1[-1]) + 5
    b2 = int(tra2[-1]) + 5
    for dt1 in dtypes:
        if dt1 == 'cl_be':
            continue
        _, _, sind1 = s.get_ell_cl(dt1, tra1, tra2, return_ind=True)
        for trb1, trb2 in trcomb[i:]:
            b3 = int(trb1[-1]) + 5
            b4 = int(trb2[-1]) + 5
            for dt2 in dtypes:
                if dt2 == 'cl_be':
                    continue
                _, _, scov2, sind2 = s.get_ell_cl(dt2, trb1, trb2, return_cov=True, return_ind=True)
                
                cov = np.load(root+'cov_new_fiducial/cov_s2222_b{}{}{}{}.npz'.format(b1, b2, b3, b4))['arr_0']
                cov = cov.reshape((lbpw.size, 4, lbpw.size, 4))
                print(tra1, tra2, trb1, trb2, dt1, dt2, b1, b2, b3, b4, ic[dt1], ic[dt2],
                      np.max(np.abs(scov[sind1[:, None], sind2] - cov[:, ic[dt1], :, ic[dt2]])))
```
